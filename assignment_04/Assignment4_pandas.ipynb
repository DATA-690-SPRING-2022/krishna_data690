## Using Pandas to read the file census_costs.txt file.

```python
df = pd.read_csv("https://raw.githubusercontent.com/wcj365/data690fall21/main/assignment_04/census_costs.txt", sep="\t", skiprows=1)
print(df)
```

### Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function


```python
def cleanse_year(x):
    a = x.replace("*","")
    return a
df["year_list"] = df["Census Year"].apply(cleanse_year)
df['year_list'] = pd.to_numeric(df['year_list'])
df

```

```python
def cleanse_pop(x):
    a = x.replace(",","")
    return a
df["pop_list"] = df["Total Population"].apply(cleanse_pop)
df['pop_list'] = pd.to_numeric(df['pop_list'])
df
```

```python
def cleanse_cost(x):
    return (x.replace(",","").replace("$","").replace(" Billion","000000000"))
df["cost_list"] = df["Census Cost"].apply(cleanse_cost)
df['cost_list'] = pd.to_numeric(df['cost_list'])
df
```

```python
def cleanse_avg(x):
    return (x.replace(",","").replace("$","").replace(" cents",""))
df["avg_list"] = df["Average Cost Per Person"].apply(cleanse_avg)
df['avg_list'] = pd.to_numeric(df['avg_list'])
df
```

## Using Pandas built-in plotting function to plot the average cost per person over time.

```python
df.plot(x='year_list', y='avg_list', figsize=(11,5), kind="area")
```
